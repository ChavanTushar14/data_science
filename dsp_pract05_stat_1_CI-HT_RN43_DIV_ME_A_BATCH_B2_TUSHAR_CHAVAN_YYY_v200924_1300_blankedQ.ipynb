## Problem Set 2: Hypothesis Testing (6 Problems)
(Run following code cell to generate problem statement)

```r
# Run this code cell to generate problem statements
if (rn==0 | grn == 0) {
    cat("\nNOT SHOWING PROBLEMS, BECAUSE YOU HAVE NOT ENTERED YOUR DETAILS")
    cat('\nEnter both your roll & gr no in the "Enter Your Information" cell above before proceeding');
    cat('\nIf you have already done so, run all cells (click Cell > Run All )')
} else {
    cat("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n")
    cat("  Start of: Problem Set 2 (6 problems): Hypothesis Testing    \n")
    cat("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n")

    sprintf("If the following are not your roll no, gr no & name, change at top under the cell titled Enter Your Information and do Cell > Run All")
    sprintf("Roll No: %s, GRN: %s, Name: %s ", rn,grn,my_name)
    cat("\n")

    set.seed(grn)
    dummy  <-  rnorm(10) # Displacing random numbers

    nq = 6 # Keep this even and also multiple of 3
    alpha_sl  <- sample(c(0.05,0.05,0.05,0.05,0.02,0.01),nq)
    outcome_options  <- c(rep('r',times=nq/2),rep('f',times=nq/2))
    outcome  <- sample(outcome_options,nq)
    side_options  <- c(rep('1r',times=nq/3),rep('1l',times=nq/3),rep('2',times=nq/3))
    side  <- sample(side_options, nq)

    xbar  <- sample(seq(200,400,by=1),nq)
    n_normal  <- sample(seq(30,100,by=1),nq/2)
    n_t  <- sample(seq(5,15,by=1),nq/2)
    n  <- c(n_normal,n_t)
    normal_or_t = c(rep('n',times=nq/2),rep('t',times=nq/2))
    sd  <- sample(30:100,nq)

    mu0  <- c()
    for (i in 1:nq) {
        if (outcome[i] %in% 'r') {
            p1 = 0.1*alpha_sl[i]
            p2 = 0.9*alpha_sl[i]
        } else { # fail to reject
            p1 = 1.1*alpha_sl[i]
            p2 = 2.0*alpha_sl[i]        
        }
        if (side[i] %in% '1l') {
            if (normal_or_t[i] %in% 'n') {
                z1 = qnorm(p1)
                z2 = qnorm(p2)
            } else {
                z1 = qt(p1,df = n[i]-1)
                z2 = qt(p2,df = n[i]-1)                
            }
        } else if (side[i] %in% '1r') {
            if (normal_or_t[i] %in% 'n') {
                z1 = qnorm(p1,lower.tail=FALSE)
                z2 = qnorm(p2,lower.tail=FALSE)
            } else {
                z1 = qt(p1,df = n[i]-1,lower.tail=FALSE)
                z2 = qt(p2,df = n[i]-1,lower.tail=FALSE)                
            }            
        } else {
            if (normal_or_t[i] %in% 'n') {
                z1 = qnorm(p1/2,lower.tail=FALSE)
                z2 = qnorm(p2/2,lower.tail=FALSE)
            } else {
                z1 = qt(p1/2,df = n[i]-1)
                z2 = qt(p2/2,df = n[i]-1)                
            }                        
        }

        me1  <-z1*sd[i]/sqrt(n[i])
        me2  <-z2*sd[i]/sqrt(n[i])
        mu01 = xbar[i] - me1
        mu02 = xbar[i] - me2
        mu0_this = sample(seq(from=mu01,to=mu02,length.out=20),1)
        mu0  <- c(mu0,round(mu0_this,1))
    }

    qn1  <- sample(1:3,3); qn2  <- sample(4:6,3)
    qn  <- c(qn1,qn2)
    qn_displ = 1
    for (i in qn){
        # Write questions
        if (normal_or_t[i] %in% 'n') {
            s_norm = "Assume normality."
        } else {
            s_norm = ""
        }
        if (side[i] %in% '1r') {
            s_check4 = "more than"
        } else if (side[i] %in% '1l') {
            s_check4 = "less than"
        } else {
            s_check4 = "different from"
        }
        cat("Q.",qn_displ,") ")
        qn_displ  <- qn_displ + 1
        cat("A sample taken from a population has a")
        cat(" size of",paste(n[i]))
        cat(", mean of",paste(xbar[i]))
        cat(" and standard deviation of",paste(sd[i]),".")
        #     cat("\nThe s")
        cat("\nDo hypotheis testing using ")
        cat("a significance level of", paste(alpha_sl[i]))
        cat(" to check if the population mean is", paste(s_check4))
        cat(" a value of", paste(mu0[i]),". ")
        cat(s_norm)
        cat("\n\n")
    }
    cat("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n")
    cat("  End of: Problem Set 2 (6 problems): Hypothesis Testing      \n")
    cat("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n")
}
```

## Problemset 2: Multiple Linear Regression
(Run following code cell to generate problem statement)

```r
# Run this code cell to generate problem statements
if (rn==0 | grn == 0) {
    cat("\nNOT SHOWING PROBLEMS, BECAUSE YOU HAVE NOT ENTERED YOUR DETAILS")
    cat('\nEnter both your roll & gr no in the "Enter Your Information" cell above before proceeding');
    cat('\nIf you have already done so, run all cells (click Cell > Run All )')
} else {
    cat("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n")
    cat("  Start of: Problemset 2: Multiple Linear Regression\n")
    cat("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n")

    cat("If the following are not your roll no, gr no & name, change at top under the cell titled Enter Your Information and do Cell > Run All")
    cat("\n\nRoll No:", rn, "GRN: ",grn, "Name: ", my_name)
    cat("\n\n")
    set.seed(grn)

    # Dataset gen
    m_trial  <-  200
    err_var  <-  50
    p_feature  <- 5
    beta0  <- runif(1,80,120)
    beta  <- round(runif(p_feature, 0.8,1.2),2)
    # Make 2 randomly chosen beta values zero
    no_ind  <- sample(1:3,1)
    ind  <- sample(1:5,no_ind)
    beta[ind]  <- 0
    x  <- c()
    for (i in 1:p_feature) {
        x  <- cbind(x,round(runif(m_trial,0,100),1))
    }
    y = beta0  + x%*%beta
    y  <- y + rnorm(m_trial,0,err_var)
    y  <- round(y,1)
    df_trial_mlr <- data.frame(feature = x, y=y)

    cat("\nPS2-Q1)")
    cat("\nGiven dataframe: df_trial_mlr with",paste(m_trial),"records")
    cat("\nCarry out linear regression of y vs all other columns")
    cat("\nusing the lm(.) function")
    cat("\nand answer the following questions:")
    cat("\n(Use a significance level of 0.05)")
    cat("\n  a) If y appears to be independent of some features, list these.")
    cat('\n     (We will do more elaborate methods for this later.')
    cat('\n      Those interested can look up subset selection,')
    cat('\n      cross-validation etc.)')
    cat("\n  b) % variance explained by the MLR fit is?")
    cat("\n  c) Consider the null hypothesis H0:")
    cat("\n     H0: y is not dependent on any of the features.")
    cat("\n     Does the fit suggest rejection of H0?")
    cat("\n  d) Residual Standard Error as % of mean of y is?")
    cat("\n  e) Find approx 95 % confidence intervals of all coefficients")
    cat("\n     (Hint: Use a critical z-score of 2, which is ok
        since at the large df here, t-distribution 
        approximates to normal.)")

    cat("\n\n")
    
    cat("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n")
    cat("  End of: Problemset 2: Multiple Linear Regression\n")
    cat("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++\n")
}
```

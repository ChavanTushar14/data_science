### Aesthetic Mappings (aes)

#### What are aesthetics ?

**Aesthetics:**  
- Visual properties of objects in the plot
- eg: In scatter plot:
    + size, shape, color, transparency of a point
    + Location of a point,x & y, are also aesthetics!
- A variable in data can be **'mapped' to an 'aesthetic'**
    + using the 'aes(.)' function
    
**Example:ggplot layer command to map class to color aesthetic:**
- The command:
      geom_point(aes(x=displ, y=hwy, color = class))  
- A different color assigned to each of the following unique class values:  
  'compact''midsize''suv''2seater''minivan''pickup''subcompact'
- A legend also displayed 
    + ('x,y' aesthetics are special: coordinate axes created for them.       So axis lines are like legends!)
       
- Terms:  
    + 'levels': values of aesthetic properties (eg color here)
    + 'scaling': process of assigning a unique level to 
       each unique variable value (eg color to car class here)

#### Mapping car 'class' to an aesthetic (SUV, minivan etc)

```r
library(tidyverse)
library(repr);options(repr.plot.width=6, repr.plot.height=3)

# ---------- Color according to class -------------------------
ggplot(data=mpg) + geom_point(aes(x=displ, y=hwy, color = class))
# Try out above with a) size = class b) alpha = class c) shape  = class or combinations
# Try out coloring for cyl >= 7
# ggplot(data=mpg) + geom_point(aes(x=displ, y=hwy, color = cyl>=7))
```

#### Manually specifying aesthetics

##### Details (Manually specifying aesthetics)

- Specify aesthetics outside the aes(.) function
- **Colors:** 
    + Hundreds of names, eg "chocolate4"
    + RGB value in hex, eg "#ff0000" is red
- **Shape:**
    + Any letter (eg "Q") or "NA" or "."
    + or numbers [0,25] giving following shapes




##### Code (Manually specifying aesthetics)

```r
library(tidyverse)
library(repr);options(repr.plot.width=6, repr.plot.height=3)

### Specify aesthetics OUTSIDE aes function
# =============== Color =========================================
ggplot(data=mpg) + 
geom_point(aes(x=displ, y=hwy), shape=18,color = "chocolate4")
# ggplot(data=mpg) + geom_point(aes(x=displ, y=hwy), color = "#ff0000")
# =============== Shape =========================================
# 0-25
# ggplot(data=mpg) + geom_point(aes(x=displ, y=hwy), shape = 4)
# Also try with size, alpha
```

### Plot of Statistical Functions: pdf / cdf

#### Normal Distribution: Probability Density Function (pdf)

```r
library(tidyverse)
library(repr);options(repr.plot.width=6, repr.plot.height=3)


x <- data.frame(x = c(-5,15))
ggplot(x) + 
stat_function(aes(x=x,color="2"),fun=dnorm,args=list(mean=5,sd=2)) +
stat_function(aes(x=x,color="1"),fun=dnorm,args=list(mean=5,sd=1)) +
stat_function(aes(x=x,color="3"),fun=dnorm,args=list(mean=5,sd=3)) +
labs(title="Normal Distribution, mean = 5")+
scale_colour_manual("sd",values=c("red","blue","orange"))
```

```r
library(tidyverse)
library(repr); options(repr.plot.widht=3, repr.plot.height=3)
x <- data.frame(x=c(1,10))
ggplot(x) + 
stat_function(aes(x=x,color="2"),fun=dnorm,args=list(mean=7,sd=2)) +
stat_function(aes(x=x,color="1"),fun=dnorm,args=list(mean=7,sd=1)) +
stat_function(aes(x=x,color="3"),fun=dnorm,args=list(mean=7,sd=3)) +
labs(title="Normal Distribution, mean=7")+
scale_colour_manual("sd", values=c("red","orange","green"))
```

#### Normal Distribution: Cumulative Distribution Function (cdf)

```r
library(tidyverse)
library(repr);options(repr.plot.width=6, repr.plot.height=3)

x <- data.frame(x = c(-8,8))
ggplot(x) + 
stat_function(aes(x=x,color="2"),fun=pnorm,args=list(mean=0,sd=2)) +
stat_function(aes(x=x,color="1"),fun=pnorm,args=list(mean=0,sd=1)) +
stat_function(aes(x=x,color="3"),fun=pnorm,args=list(mean=0,sd=3)) +
labs(title="Cumulative Distribution Function: Normal Distribution, mean = 0")+
scale_colour_manual("sd",values=c("red","blue","orange")) +
geom_hline(aes(yintercept=pnorm(-1)),color = 'green') +
geom_hline(aes(yintercept=pnorm(1)),color = 'green') +
scale_x_continuous(breaks=seq(-8,8,by=2)) 
# scale_y_manual(values=c(pnorm(-1),pnorm(1)))
```

```r
library(tidyverse)
library(repr);options(repr.plot.width=6, repr.plot.height=4)

x <- data.frame(x = c(-10,10))
ggplot(x) + 
stat_function(aes(x=x,color="2"),fun=pnorm,args=list(mean=2,sd=2)) +
stat_function(aes(x=x,color="1"),fun=pnorm,args=list(mean=2,sd=1)) +
stat_function(aes(x=x,color="3"),fun=pnorm,args=list(mean=2,sd=3)) +
labs(title="Cumulative Distribution Function: Normal Distribution, mean = 2")+
scale_colour_manual("sd",values=c("red","blue","orange")) +
geom_hline(aes(yintercept=pnorm(-1)),color = 'darkred') +
geom_hline(aes(yintercept=pnorm(1)),color = 'darkred') +
scale_x_continuous(breaks=seq(-10,10,by=2)) 
```
